```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("house_data.csv")
print(df.head())
```

```python
df = df.dropna()
X = df[['area', 'bedrooms', 'bathrooms', 'age']]
y = df['price']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
pred = model.predict([[1200, 3, 2, 5]])
print(pred)
```

```python
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```
